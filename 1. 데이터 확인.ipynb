# 항만 혼잡도 반환 함수
- 전달 인자: port_name, date
- 사용 데이터셋: port_congestion
  - 컬럼: 'PORT_CD', 'VSL_ID', 'ATA_DT', 'ATB_DT', 'ATD_DT', 'working_time'

```python
def congestion_rate(port_name, date):
    # port_df 생성
    port_df = port_congestion[port_congestion['PORT_CD'] == port_name]

    # 혼잡도 counting에 필요한 열 추가
    port_df = port_df.sort_values('ATB_DT')
    port_df['ATB_date'] = port_df['ATB_DT'].apply(lambda x: f'{x.year}-{x.month:02}-{x.day:02}')
    port_df['ATD_date'] = port_df['ATD_DT'].apply(lambda x: f'{x.year}-{x.month:02}-{x.day:02}')

    port_df['work_start_time'] = port_df['ATB_date'] + port_df['ATB_DT'].apply(lambda x: f' {x.hour:02}:00:00')
    port_df['work_end_time'] = port_df['ATD_date'] + port_df['ATD_DT'].apply(lambda x: f' {x.hour:02}:00:00')

    port_df.index = range(len(port_df))

    port_df['worked_time_list'] = pd.Series([pd.date_range(start=port_df['work_start_time'].iloc[i],end=port_df['work_end_time'].iloc[i], freq='H').tolist() for i in range(len(port_df))])

    # hourly port congestion 카운팅
    date_timestamp = pd.date_range(start=date+' '+'00:00:00', end=date+' '+'23:00:00', freq='H')
    hourly_port_congestion_df = pd.DataFrame(index=date_timestamp, columns=['working_vessel_cnt'])

    # working_vessel_cnt 값 연산해서 데이터프레임에 채워넣기
    for time in date_timestamp:
        cnt=0
        for time_list in port_df['worked_time_list'].values:
            cnt+=time_list.count(time)
            hourly_port_congestion_df.loc[time] = cnt

    return hourly_port_congestion_df # 해당 항만 - 일자 - 시간대 별 혼잡도 df
```

### working time 이상치 제거 함수

```python
q1 = port_workingtime['working_time'].quantile(0.25)
q3 = port_workingtime['working_time'].quantile(0.75)
iqr = q3 - q1
```

```python
# df: port_workingtime; port와 workingtime 열만 있는 그래프
def delete_outlier(port, df):
    port_df = df[df['PORT_CD'] == port]
    
    q1 = port_df['working_time'].quantile(0.25)
    q3 = port_df['working_time'].quantile(0.75)
    iqr = q3 - q1
    min = q1 - 1.5*iqr
    max = q3 + 1.5*iqr

    port_df['outlier'] = port_df['working_time'].apply(lambda x: True if not min <= x <= max else False)
    port_df = port_df[port_df['outlier'] == False][['PORT_CD', 'working_time']]

    return port_df
```

```python
port_workingtime['PORT_CD'].unique()
```

```python
# port별 이상치 제거된 workingtime df
JPTYO_df = delete_outlier('JPTYO', port_workingtime)
HKHKG_df = delete_outlier('HKHKG', port_workingtime)
SGSIN_df = delete_outlier('SGSIN', port_workingtime)
CNSHA_df = delete_outlier('CNSHA', port_workingtime)
KRINC_df = delete_outlier('KRINC', port_workingtime)
JPOSA_df = delete_outlier('JPOSA', port_workingtime)
KRPUS_df = delete_outlier('KRPUS', port_workingtime)

port_workingtime_trim = pd.concat([JPTYO_df, HKHKG_df, SGSIN_df, CNSHA_df, KRINC_df, JPOSA_df, KRPUS_df])
```

```python
# 제거된 이상치 개수 확인
for port_name in port_workingtime['PORT_CD'].unique():
    print(f'{port_name}항')
    before = len(port_workingtime[port_workingtime["PORT_CD"]== port_name])
    after = len(delete_outlier(port_name, port_workingtime))
    print(f'이상치 제거 전: {before}개')
    print(f'이상치 제거 후: {after}개')
    print(f'제거된 이상치: {before - after}개')
    print()
```

### 이상치 제거

```python
# df: shiptype - working time 데이터 프레임
def delete_outlier_shiptype(shiptype, df):
    shiptype_df = df[df['SHIP_TYPE'] == shiptype]
    
    q1 = shiptype_df['working_time'].quantile(0.25)
    q3 = shiptype_df['working_time'].quantile(0.75)
    iqr = q3 - q1
    min = q1 - 1.5*iqr
    max = q3 + 1.5*iqr

    shiptype_df['outlier'] = shiptype_df['working_time'].apply(lambda x: True if not min <= x <= max else False)
    shiptype_df = shiptype_df[shiptype_df['outlier'] == False][['SHIP_TYPE', 'working_time']]

    return shiptype_df
```

```python
shiptype_workingtime['SHIP_TYPE'].unique()
```

```python
# 이상치 제거
CONTAINER_df = delete_outlier_shiptype('CONTAINER', shiptype_workingtime)
GENERAL_CARGO_df = delete_outlier_shiptype('GENERAL_CARGO', shiptype_workingtime)
DRY_BULK_df = delete_outlier_shiptype('DRY_BULK', shiptype_workingtime)
ROLL_ON_ROLL_OFF_df = delete_outlier_shiptype('ROLL_ON_ROLL_OFF', shiptype_workingtime)
SPECIAL_CRAFT_df = delete_outlier_shiptype('SPECIAL_CRAFT', shiptype_workingtime)
LNG_CARRIER_df = delete_outlier_shiptype('LNG_CARRIER', shiptype_workingtime)
VEHICLE_PASSENGER_df = delete_outlier_shiptype('VEHICLE_PASSENGER', shiptype_workingtime)
TUG_df = delete_outlier_shiptype('TUG', shiptype_workingtime)
TANKER_CHEMICALS_df = delete_outlier_shiptype('TANKER_CHEMICALS', shiptype_workingtime)
OTHER_CHEMICALS_df = delete_outlier_shiptype('OTHER', shiptype_workingtime)
GENERAL_TANKER_df = delete_outlier_shiptype('GENERAL_TANKER', shiptype_workingtime)
TANKER_PRODUCT_df = delete_outlier_shiptype('TANKER_PRODUCT', shiptype_workingtime)

shiptype_workingtime_trim = pd.concat([CONTAINER_df, GENERAL_CARGO_df, DRY_BULK_df, ROLL_ON_ROLL_OFF_df,
                                      SPECIAL_CRAFT_df, LNG_CARRIER_df, VEHICLE_PASSENGER_df, TUG_df,
                                      TANKER_CHEMICALS_df, OTHER_CHEMICALS_df, GENERAL_TANKER_df, TANKER_PRODUCT_df])
```

```python
# 제거된 이상치 개수 확인
for shiptype in shiptype_workingtime['SHIP_TYPE'].unique():
    print(f'{shiptype} 선')
    before = len(shiptype_workingtime[shiptype_workingtime["SHIP_TYPE"]== shiptype])
    after = len(delete_outlier_shiptype(shiptype, shiptype_workingtime))
    print(f'이상치 제거 전: {before}개')
    print(f'이상치 제거 후: {after}개')
    print(f'제거된 이상치: {before - after}개')
    print()
```
